```python
f = open("accs_2_layers_out.txt", "r")
text = f.readline()
f.close()

acc_2_layers = []
accs = text.split(',')
for acc in accs:
    acc_2_layers.append(float(acc))
```

```python
f = open("accs_1_layer_out.txt", "r")
text = f.readline()
f.close()

acc_1_layer = []
accs = text.split(',')
for acc in accs:
    acc_1_layer.append(float(acc))
```

```python
print(acc_2_layers)
```

```python
print(acc_1_layer)
```

```python
import matplotlib.pyplot as plt

x1 = acc_1_layer
x2 = acc_2_layers
y = range(1, len(acc_1_layer) + 1)
    
plt.plot(y, x1, 'g', label='1_layer_acc')
plt.plot(y, x2, 'r', label='2_layers_acc')

plt.title("Accuracy vs Generation")
plt.xlabel("generation")
plt.ylabel("accuracy")

plt.legend()
plt.show()
```
